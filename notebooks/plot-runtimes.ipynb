```python
import re
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scienceplots

plt.style.use('science')
```

## LNS runtime iteration

```python
data = pd.read_csv("../data/runtime-lns-iteration.csv")
```

```python
colors = ["#88CCEE", "#CC6677"]
linestyles = ["--", "-."]
```

```python
fig, ax = plt.subplots(1, figsize=(4, 4), dpi=400)


for serv in [0, 1]:
    tmp = data[data['serv'] == serv]
    ax.plot(tmp['n'], tmp['time/iter'], label=f"{'High' if serv else 'Low'} SCV", marker='.', color=colors[serv], linestyle=linestyles[serv])

# ax.set_yscale('log')  # Set y-axis to be logarithmic scale
ax.set_ylabel("Time (s)")
ax.set_xlabel("Instance size $n$")

plt.legend()
plt.tight_layout()
plt.savefig("../figures/time-per-iteration.pdf")
```

## Optimal schedule
